```python
from random import randint
import time

Board = ['Go', 'Mediterranean Avenue', 'Community Chest', 'Baltic Avenue', 'Income tax', 'Reading Railroad', 'Oriental Avenue', 'Chance', 'Vermont Avenue', 'Connecticut Avenue', 'Jail/Just Visiting', 'St Charles Place', 'Electric Company', 'States Avenue', 'Virginia Avenue', 'Pennsylvania Railroad', 'St James Place', 'Community Chest', 'Tennessee Avenue', 'New York Avenue', 'Free Parking', 'Kentucky Avenue', 'Chance', 'Indiana Avenue', 'Illinois Avenue', 'B & O Railroad', 'Atlantic Avenue', 'Ventnor Avenue', 'Water Works', 'Marvin Gardens', 'Go To Jail', 'Pacific Avenue', 'North Carolina Avenue', 'Community Chest', 'Pennsylvania Avenue', 'Short Line Railroad', 'Chance', 'Park Place', 'Luxury Tax', 'Boadwalk']
BoardPrices = [200, 60, 0, 60, 200, 200, 100, 0, 100, 120, 0, 140, 150, 140, 160, 200, 180, 0, 180, 200, 0, 220, 0, 220, 240, 200, 260, 260, 150, 280, 50, 300, 300, 0, 320, 200, 0, 350, 100, 400]
BuyableProperties = ['Mediterranean Avenue', 'Baltic Avenue', 'Reading Railroad', 'Oriental Avenue', 'Vermont Avenue', 'Connecticut Avenue', 'St Charles Place', 'Electric Company', 'States Avenue', 'Virginia Avenue', 'Pennsylvania Railroad', 'St James Place', 'Tennessee Avenue', 'New York Avenue', 'Kentucky Avenue', 'Indiana Avenue', 'Illinois Avenue', 'B & O Railroad', 'Atlantic Avenue', 'Ventnor Avenue', 'Water Works', 'Marvin Gardens', 'Pacific Avenue', 'North Carolina Avenue', 'Pennsylvania Avenue', 'Short Line Railroad', 'Park Place', 'Boadwalk']
chest_chance = ['Community Chest', 'Chance']
TaxProperties = ['Income tax', 'Luxury Tax',]
RandoSpaces = ['Go', 'Jail/Just Visiting', 'Free Parking',]
start_of_game = False
location = 0
OtherPlayersProperties = []
MyProperties = []
InJail = False 
TurnsPased = 0
GetOutOfJailCard = False

def Start():
    global start_of_game 
    global Cash 
    
    while start_of_game == False:
        
        amount = input('How Much Money Do I Start With? (Default = 1500, Custom = You Enter The Amount): ')
        if (amount.lower() == 'default'):
            Cash = 1500
            print(f'I Have {Cash} Dollars')
            start_of_game = True
            time.sleep(0.2)
            Roll()
        elif (amount.lower() == 'custom'):
            amount = input('How Much Money Do I Start With? (100, 200): ')
            Cash = int(amount)
            print(f'I Have {Cash} Dollars')
            start_of_game = True
            time.sleep(0.2)
            Roll()
        else:
            start_of_game = False
            
def Roll():
    #Rolling Dices
    global location
    global spaces 
    
    spaces = randint(2,12)
    location = location + spaces
    Move()
    
def Move():
    #Moving Pice On The Board
    global location
    global spaces 
    global OtherPlayersProperties
    global MyProperties
    global Cash 
    global InJail
    global TurnsPased
    global GetOutOfJailCard
    
    # If He is Passing Go
    if (location > 40):
        location = 0 
        location = location + spaces
        Cash = Cash + 200
        print(f'My New Balance Is {Cash}')
        print(f'I Am On {Board[location]}')
        
    elif (location < 40):
        print(f'I Am On {Board[location]}')
    
    # If He Goes To Jail
    if (Board[location] == 'Go To Jail'):
        location = 10
        InJail = True
        OtherPlayersTurn()
    #Geting Out Of Jail    
    if (location == 10 and InJail == True):
        roll = randint(2,12)
        if Cash > 50:
            buyoutchance = randint(0,100)
            if buyoutchance <= 80:
                Cash = Cash - 50 
                print(f'Here Is 50 Dollars My New Balance Is {Cash}')
                TrunsPased = 0 
                InJail = False 
        elif TurnsPased == 3:
            InJail = False
            TrunsPased = 0
        elif (roll == 12):
            InJail = False
            TrunsPased = 0
        elif GetOutOfJailCard == True:
            GetOutOfJailCard = False
            InJail = False
            TrunsPased = 0
        else:
            InJail = True
            OtherPlayersTurn()    
    # Comunity Chest / Chance            
    if (Board[location] in chest_chance):
        WhatHappens = input('Do I Move To A Different Square?(Move) Do I Get Money?(You Get Money) Do I Lose Money? (You Lose Money) Do I Get A Get Out Of Jail Free Card? (Get Out Of Jail Card) Am I Sent To Jail? (Go To Jail): ')
        if (WhatHappens.lower() == 'you get money'):
            amount = input('How Much Money Do I Get? (50/100): ')
            Cash = Cash + int(amount)
            print(f'My New Balance Is {Cash}')
            OtherPlayersTurn()
        elif (WhatHappens.lower() == 'you lose money'):
            amount = input('How Much Money Do I Lose? (50/100): ')
            Cash = Cash - int(amount)
            print(f'My New Balance Is {Cash}')
            OtherPlayersTurn()
        elif (WhatHappens.lower() == 'go to jail'):
            print('Reeeeee Okay ')
            InJail = True 
            location = 10
            OtherPlayersTurn()
        elif (WhatHappens.lower() == 'get out of jail card'):
            GetOutOfJailCard = True
            OtherPlayersTurn()
        elif (WhatHappens.lower() == 'move'):
            loc = input('What Square Do I Go To Start At Go And Count Go As 0 Then 1 2 3... ')
            location = int(loc)
            print(f'I Am At {Board[location]}')
            OtherPlayersTurn()
    #Tax 
    if (Board[location] in TaxProperties):
        amount = BoardPrices[location]
        Cash = Cash - int(amount)
        print(f'Nooooooooo Fine Heres {amount} Dollars My New Balance Is {Cash} Dollars')
        OtherPlayersTurn()    
    
    #FreeSpaces 
    if (Board[location] in RandoSpaces):
        OtherPlayersTurn()
    #If He Knows The Propertie Is Taken
    if (Board[location] in OtherPlayersProperties):
            pay = input('How Much Money Do I Own You? (50/100)')
            Cash = Cash - int(pay)
            print(f'Heres {pay} My New Balance Is {Cash} Dollars')
            OtherPlayersTurn()
    elif (Board[location] in MyProperties):
            OtherPlayersTurn() 
    #Finding Out If He Can Buy The Propertie
    elif (Board[location] in BuyableProperties):
            status = input('Is This Property Owned By Anyone? (Yes/No)')
            if (status.lower() == 'yes'):
                OtherPlayersProperties.append(Board[location])   
                pay = input('How Much Money Do I Own You? (50/100)')
                Cash = Cash - int(pay)
                print(f'My New Balance Is {Cash}')
                OtherPlayersTurn()
            #Buying Properties
            elif (status.lower() == 'no' and BoardPrices[location] < Cash):
                if (location >= 0 and location <= 10):
                    chanceofbuying = randint(0,100)
                    if (chanceofbuying <= 30):
                        Cash = Cash - BoardPrices[location]
                        MyProperties.append(Board[location])
                        print(f'Here Is {BoardPrices[location]} Dollars I Am Now The New Proud Owner Of {Board[location]}')
                        print(f'My New Balance Is {Cash}')
                        OtherPlayersTurn()
                    else:
                        print('Okay I Dont Think I Want To Buy That Property Right Now Tho')
                        OtherPlayersTurn()
                if (location >= 11 and location <= 21):
                    chanceofbuying = randint(0,100)
                    if (chanceofbuying <= 75):
                        Cash = Cash - BoardPrices[location]
                        MyProperties.append(Board[location])
                        print(f'Here Is {BoardPrices[location]} Dollars I Am Now The New Proud Owner Of {Board[location]}')
                        print(f'My New Balance Is {Cash}')
                        OtherPlayersTurn()
                    else:
                        print('Okay I Dont Think I Want To Buy That Property Right Now Tho') 
                        OtherPlayersTurn()
                if (location >= 22 and location <= 32):
                        chanceofbuying = randint(0,100)
                        if (chanceofbuying <= 75):
                            Cash = Cash - BoardPrices[location]
                            MyProperties.append(Board[location])
                            print(f'Here Is {BoardPrices[location]} Dollars I Am Now The New Proud Owner Of {Board[location]}')
                            print(f'My New Balance Is {Cash}')
                            OtherPlayersTurn()
                        else:
                            print('Okay I Dont Think I Want To Buy That Property Right Now Tho') 
                            OtherPlayersTurn()
                if (location >= 33 and location <= 40):
                        chanceofbuying = randint(0,100)
                        if (chanceofbuying <= 20):
                            Cash = Cash - BoardPrices[location]
                            MyProperties.append(Board[location])
                            print(f'Here Is {BoardPrices[location]} Dollars I Am Now The New Proud Owner Of {Board[location]}')
                            print(f'My New Balance Is {Cash}')
                            OtherPlayersTurn()
                        else:
                            print('Okay I Dont Think I Want To Buy That Property Right Now Tho')   
                            OtherPlayersTurn()
                            
def OtherPlayersTurn():
    global TurnsPased
    global InJail
    global Cash
    
    if Cash <= 0:
        print('Good Game Im a Broke Boy')
        EndGame()
        
    elif Cash > 0:  
        MyTurn = input('Is It My Turn? (Yes/No): ')
        
        if (MyTurn.lower() == 'no' and InJail == True):
            TurnsPased += 1 
            
        else:
            payme = input(f'Did Any One Land On One Of My Properties? {MyProperties} (Yes/No)')
            
        if payme.lower() == 'yes':
            payme = input('Lets Go Pay Me The Amount On The Card (50/100): ')
            Cash = Cash+int(payme)
            print(f'Yes Sir My New Balance Is {Cash}')
                
        if MyTurn == 'yes':
            Roll()
        else:
            OtherPlayersTurn()
def EndGame():
    None  
Start()
```
